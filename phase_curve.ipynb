```python
import json
import glob
import os

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import utils
from add_fourier import get_fourier_fn
```

```python
data_file = 'pluto_435'
with open(f"output/{data_file}.json", 'r') as f:
    data = json.load(f)

df = pd.DataFrame.from_dict(data['data'])
df.sort_values('lon', inplace=True)

plt.title("Pluto F435W")

plt.scatter(df.lon, df.converted_vegamag_NH_RED, marker='o', c='red', label='NH Red')
plt.scatter(df.lon, df.converted_vegamag_NH_BLUE, marker='+', c='blue', label='NH Blue')
plt.scatter(df.lon, df.multi_vegamag, marker='x', c='green', label='Combined')

buie_x = np.arange(360)
buie_y = get_fourier_fn(f"data/buie_fourier_{data_file}.json")(buie_x)
plt.plot(buie_x, buie_y, c="black", label="Buie Fourier Fit")

plt.legend()
plt.gca().invert_yaxis()
plt.xlabel("Longitude (deg)")
plt.xlabel("Flux (Vegamag)")
figname = f"figures/{data_file}.eps"
plt.savefig(figname, bbox_inches='tight')
plt.show()
```
