```python
import json
import glob
import os

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import utils
```

```python
data_file = 'output/pluto_435.json'
with open(data_file, 'r') as f:
    data = json.load(f)

df = pd.DataFrame.from_dict(data['data'])
df.sort_values('lon', inplace=True)

plt.title("Pluto F435W")

plt.scatter(df.lon, df.converted_vegamag_NH_RED, marker='o', c='red', label='NH Red')
plt.scatter(df.lon, df.converted_vegamag_NH_BLUE, marker='+', c='blue', label='NH Blue')
plt.scatter(df.lon, df.multi_vegamag, marker='x', c='green', label='Combined')
plt.plot(df.lon, df.fourier_vegamag, c='black', label='Buie Fourier')

plt.legend()
plt.gca().invert_yaxis()
figname = os.path.basename(data_file).replace('.json', '.jpg')
plt.savefig(figname, bbox_inches='tight')
plt.show()
```

```python
data_file = 'output/pluto_435.json'
with open(data_file, 'r') as f:
    data = json.load(f)

df = pd.DataFrame.from_dict(data['data'])
df.sort_values('phase', inplace=True)

plt.title("Pluto F435W delta")

plt.scatter(df.phase, df.multi_vegamag_delta)

# Fits
fit = np.polyfit(df.phase, df.multi_vegamag_delta, 1)
x = [df.phase.min(), df.phase.max()]
plt.plot(x, np.poly1d(fit)(x), '--')

plt.gca().set_ylim([-.1,0.2])
# plt.legend()
plt.gca().invert_yaxis()
figname = os.path.basename(data_file).replace('.json', '.jpg')
plt.savefig(figname, bbox_inches='tight')
plt.show()
```
